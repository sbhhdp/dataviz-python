```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
%matplotlib inline

df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')

df1 = df[df['Date'].between('2005-01-01', '2014-12-31')]
df2 = df[df['Date'].between('2015-01-01', '2015-12-31')]

#df['Date'] = df['Date'].astype('datetime64[ns]')
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])

df1['day of the year'] = df1['Date'].apply(lambda x : x.strftime('%m-%d'))
df2['day of the year'] = df2['Date'].apply(lambda x : x.strftime('%m-%d'))
#df['day of the year']

df1.set_index('Date', inplace = True)
df2.set_index('Date', inplace = True)
#df.sort_index(inplace = True)

df1.drop(['2008-02-29', '2012-02-29'], inplace = True)

gp = df1.groupby('day of the year')
gp1 = df2.groupby('day of the year')

tmax = gp.max()
tmin = gp.min()

tmax['day no'] = list(range(1, 366))
tmin['day no'] = list(range(1, 366))

tmax1 = gp1.max()
tmin1 = gp1.min()

tmax1['day no'] = list(range(1, 366))
tmin1['day no'] = list(range(1, 366))

mask1 = tmax1['Data_Value'] > tmax['Data_Value']
mask2 = tmin1['Data_Value'] < tmin['Data_Value']

rb1 = tmax1[mask1]
rb2 = tmin1[mask2]



x = np.arange(1, 366)

pos = [(365/12)*i for i in range(12)]
months = ['Jan 1', 'Feb 1', 'Mar 1', 'Apr 1', 'May 1', 'Jun 1', 'July 1', 'Aug 1', 'Sep 1', 'Oct 1', 'Nov 1', 'Dec 1']

plt.figure(figsize = (20, 10))
plt.plot(x, tmax['Data_Value'], color = '#e63946', label = 'Max Temp')
plt.plot(x, tmin['Data_Value'], color = '#2a9d8f', label = 'Min Temp')
plt.scatter(rb1['day no'], rb1['Data_Value'], s = 30, lw = 1, color = '#6a040f', label = 'Record High in 2015')
plt.scatter(rb2['day no'], rb2['Data_Value'], s = 30,lw = 1, color = '#000000', label = 'Record Low in 2015')
plt.gca().fill_between(x, tmax['Data_Value'], tmin['Data_Value'], facecolor = '#d8f3dc', alpha = 0.5)
#plt.xlabel('Day Of The Year')
plt.ylabel('Temp (tenth of a Centigrade)', fontsize = 14)

plt.xticks(pos, months, fontsize = 14)
plt.title('Average Observed Temperatures on a day from 2005 to 2014', fontsize = 14)
plt.tick_params(bottom = False, left = False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.legend()
#plt.savefig('image.png')
plt.show()
```
