# Histograms

```python
np.random.normal(0, 1, 10)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]
for n in range(len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(0, 1, sample_size)
    axs[n].hist(sample)
    axs[n].set_title(f'sample size {sample_size}')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]

for n in range(len(axs)):
    sample_size = 10 ** (n+1)
    sample = np.random.normal(0, 1, sample_size)
    axs[n].hist(sample, bins = 100)
    axs[n].set_title(f'n = {sample_size}')
    
    
```

```python
fig, ax = plt.subplots(2, 2, sharex = True)
for n in range(len(ax.flatten())):
    sample_size = 10**(n+1)
    sample = np.random.normal(0, 1, sample_size)
    ax.flatten()[n].hist(sample)
    ax.flatten()[n].set_title(f'n = {sample_size}')
    
    
```

```python
fig, ax = plt.subplots(2, 2, sharex = True)
for n in range(len(ax.flatten())):
    sample_size = 10**(n+1)
    sample  = np.random.normal(0, 1, sample_size)
    ax.flatten()[n].hist(sample, bins = 15)
    
```

```python
import matplotlib.gridspec as gridspec
```

```python
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])
```

```python
y = np.random.normal(0, 1, 10000)
x = np.random.random(10000)

top_histogram.hist(x, bins = 100)
s = side_histogram.hist(y, bins = 100, orientation = 'horizontal')
lower_right.scatter(x, y)
```

```python
top_histogram.clear()
top_histogram.hist(y, density=True, bins = 100)
side_histogram.clear()
side_histogram.hist(x, density=True, bins = 100, orientation = 'horizontal')
side_histogram.invert_xaxis()
```

```python
for ax in [top_histogram, lower_right]:
    ax.set_xlim(-5, 5)
    
for ax in [side_histogram, lower_right]:
    ax.set_ylim(0, 1)
```

# Box And Whisker Plots

```python
import pandas as pd
```

```python
normal_sample = np.random.normal(0.0, 1, 10000)
random_sample = np.random.random(10000)
gamma_sample = np.random.gamma(2, size = 10000)

df = pd.DataFrame({'normal': normal_sample, 'random' : random_sample, 'gamma' : gamma_sample})
```

```python
df
```

```python
df.describe()
```

```python
plt.figure()

_ = plt.boxplot(df['normal'], whis = 'range')
```

```python
plt.clf()

_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis = 'range')
```

```python
plt.figure()
_ = plt.hist(df['gamma'], bins = 100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis = 'range')
ax2 = mpl_il.inset_axes(plt.gca(), height = '40%', width = '60%', loc = 2)
ax2.hist(df['gamma'], bins = 100)
ax2.margins(x = 0.5)
```

```python
ax2.yaxis.tick_right()
```

# Animation

```python
import matplotlib.animation as animation

n = 1000
x = np.random.randn(n)
```

```python
def update(curr):
    if curr  == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins = bins)
    plt.axis([-4, 4, 0, 200])
    plt.gca().set_title('Sampling the normal distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate(f'n = {curr}', [3, 27])
    
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval = 50)
```
